```python
import numpy as np
import lmbh
```

```python
n = 5

rng = np.random.default_rng(seed=42)
sim = lmbh.Simulation(
    graph = lmbh.GraphGenerators.ba_graph_of_allies(n, m = 3),
    prior_params = lmbh.UniformRandomPriorParams(n=n, partisans=[-1]),
    coin_params = lmbh.RandomCoinParams(),
)
```

```python
sim.analyse.show_graph()
```

```python
sim.analyse.plot_all_mean()
```

```python
sim.analyse.plot_asymptotic_learning_steps()
```

```python
sim.analyse.plot_distr(-1)
```
